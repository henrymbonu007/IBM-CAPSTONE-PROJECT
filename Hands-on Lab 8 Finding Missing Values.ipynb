### 3. Finding Missing Values
##### Task 2: Identify missing values for all columns.


```python
## Write your code here
# Display missing values for all columns
missing_values = df.isnull().sum()
print(missing_values)
```

##### Task 3: Visualize missing values using a heatmap (Using seaborn library).



```python
!pip install pandas seaborn matplotlib
```

```python
## Write your code here
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Plotting the heatmap
plt.figure(figsize=(15, 6))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis', yticklabels=False)
plt.title('Missing Values Heatmap')
plt.show()
```

##### Task 4: Count the number of missing rows for a specific column (e.g., `Employment`).


```python
## Write your code here
# Count missing values in the 'Employment' column
missing_count = df['Employment'].isna().sum()
print(missing_count)
```

### 4. Imputing Missing Values
##### Task 5: Identify the most frequent (majority) value in a specific column (e.g., `Employment`).


```python
## Write your code here
# Identify the most frequent value in the 'Employment' column
most_frequent_value = df['Employment'].mode()[0]
print(most_frequent_value)
```

##### Task 6: Impute missing values in the `Employment` column with the most frequent value.



```python
## Write your code here
# Impute missing values with the most frequent value
df['Employment'].fillna(most_frequent_value, inplace=True)
```

### 5. Visualizing Imputed Data
##### Task 7: Visualize the distribution of a column after imputation (e.g., `Employment`).


```python
!pip install scikit-learn
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer

# Imputation
imputer = SimpleImputer(strategy='most_frequent')

# Define a function to shorten the labels
def shorten_labels(labels, max_length=5):
    return [label if len(label) <= max_length else label[:max_length] + '...' for label in labels]

try:
    df['Employment'] = imputer.fit_transform(df[['Employment']])
    print(df['Employment'])
except ValueError as e:
    print(f"Imputation error: {e}")

# Visualization
plt.figure(figsize=(15, 5))
employment_counts = df['Employment'].value_counts()
short_labels = shorten_labels(employment_counts.index, max_length=5)

sns.barplot(x=short_labels, y=employment_counts.values)
plt.title('Distribution of Employment After Imputation')
plt.xlabel('Employment')
plt.ylabel('Frequency')
plt.xticks(rotation=90)

plt.show()
```
